```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from pandas import Series

Football_Team = pd.read_csv("Laliga.csv",sep = ',')

#### Self Study : Analysing the data & datatypes.

Football_Team.shape
#Football_Team.head()
#Football_Team.info()
#Football_Team.dtypes
#Football_Team.describe()
#Football_Team.shape
#display(Football_Team)

#### ----------------------------------------------------------------------------------------------------------------

#### Observation : Found first row appears as Unnamed: 0 & so on. Also Let update the dataframe Columns index.

New_Header = Football_Team.iloc[0]
Football_Team.columns = New_Header
Football_Team = Football_Team.drop(Football_Team.index[0])
#display(Football_Team)

#### ----------------------------------------------------------------------------------------------------------------

#### Q1) Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points)

Football_Team = Football_Team.replace('-',0)
print('\n Q1) Football_Team - Replacing dashes with 0 in dataset :\n',Football_Team)
display(Football_Team)

#### ----------------------------------------------------------------------------------------------------------------

#### Q2) Print all the teams which have started playing between 1930-1980. (5 points)

Football_Team.insert(18, "Football_Team_which_have_started_playing_between_1930_to_1980","Any")

Football_Team["Football_Team_which_have_started_playing_between_1930_to_1980"]= Football_Team['Debut'].astype(str).str[:4].astype(str)

Football_Team_which_have_started_playing_between_1930_to_1980 = Football_Team[Football_Team['Football_Team_which_have_started_playing_between_1930_to_1980'].between('1930', '1980')]

print('\n Q2) Football_Team_which_have_started_playing_between_1930_to_1980 are :\n',Football_Team_which_have_started_playing_between_1930_to_1980)

display(Football_Team_which_have_started_playing_between_1930_to_1980)
Football_Team_which_have_started_playing_between_1930_to_1980.shape

#### ----------------------------------------------------------------------------------------------------------------

#### Q3) Print the list of teams which came Top 5 in terms of points (5 points)

#Football_Team.info()
Football_Team = Football_Team.astype({"Points" : int})
#Football_Team.info()


Football_Team_First_Top_Five_Based_On_Points_values = Football_Team.sort_values("Points", axis = 0, ascending=False)
Football_Team_First_Top_Five_Based_On_Points_values = Football_Team_First_Top_Five_Based_On_Points_values.iloc[:5,]
print('\n Q3) Football_Team_First_Top_Five_Based_On_Points_values are :\n',Football_Team_First_Top_Five_Based_On_Points_values)
display(Football_Team_First_Top_Five_Based_On_Points_values)

#### ----------------------------------------------------------------------------------------------------------------

#### Q4) Write a function with name Goal_diff_count which should return all the teams with their 
#### Goal  Differences. Using the same function, find the team which has maximum and minimum 
#### goal  difference. (5 points)  (Goal_diff_count = GoalsFor  GoalsAgainst ).

Football_Team["GoalsFor"] = Football_Team["GoalsFor"].astype(str).astype(int)
Football_Team["GoalsAgainst"] = Football_Team["GoalsAgainst"].astype(str).astype(int)

Football_Team.insert(8,"Goal_diff_count","Any")
Football_Team["Goal_diff_count"] = Football_Team ["GoalsFor"] - Football_Team["GoalsAgainst"]
Football_Team_Goal_diff_count = Football_Team.sort_values("Goal_diff_count", axis = 0, ascending=False)
#Football_Team_Goal_diff_count.info()
print('\n Q4) Football_Team_Goal_diff_count is :\n',Football_Team_Goal_diff_count)
display(Football_Team_Goal_diff_count)

####-----------------------------------------------------------------------------------------------------------------

#### Q5) Create a new column with name Winning Percent and append it to the data set (5 points)
#### Percentage of Winning = (GamesWon / GamesPlayed)*100
#### If there are any numerical error, replace it with 0%
#### Print the top 5 teams which has the highest Winning percentage

Football_Team.insert(4, "Winning Percent","Any")
Football_Team["GamesPlayed"] = Football_Team["GamesPlayed"].astype(str).astype(int)
Football_Team["GamesWon"] = Football_Team["GamesWon"].astype(str).astype(int)
Football_Team['Winning Percent']= (((Football_Team.GamesWon)/(Football_Team.GamesPlayed))*100)
Football_Team.fillna(0)
Top_Winning_Percent_Football_Team = Football_Team.sort_values("Winning Percent", axis = 0, ascending=False)
First_Top_Five_Winning_Percent_Football_Team = Top_Winning_Percent_Football_Team.iloc[:5,]
print('\n Q5) First_Top_Five_Winning_Percent_Football_Team_are  :\n',First_Top_Five_Winning_Percent_Football_Team)
display(First_Top_Five_Winning_Percent_Football_Team)

#### ----------------------------------------------------------------------------------------------------------------

#### Q6) Group teams based on their Best position and print the sum of their points for all positions (5 points)

Football_Team["BestPosition"] = Football_Team["BestPosition"].astype(str).astype(int)
Group_Football_Team_based_on_Best_position_Sum_Of_Points = Football_Team.groupby(["BestPosition"]).sum()
Group_Football_Team_based_on_Best_position_Sum_Of_Points = Group_Football_Team_based_on_Best_position_Sum_Of_Points.iloc[:,0:1]
print ('\n Q6) Group teams based on their Best position & Sum of Points are :\n',Group_Football_Team_based_on_Best_position_Sum_Of_Points)
display(Group_Football_Team_based_on_Best_position_Sum_Of_Points)

#### ----------------------------------------------------------------------------------------------------------------
```

### Load tensorflow

```python
import tensorflow as tf
import numpy as np
```

### Collect Data

We will use MNIST dataset for this exercise. This dataset contains images of hand written numbers with each image being a black & white picture of size 28x28. We will download the data using tensorflow API. The dataset has 60,000 training examples and 10,000 test examples. Please note that images have already been converted to numpy arrays.

```python
#@title
# Download datasets. 
(trainX, trainY),(testX, testY) = tf.keras.datasets.mnist.load_data()
```

### Convert Output label to multiple values

```python
# Check current label size....
testY[0]
```

```python
# Convert labels to one hot encoding.
trainY = tf.keras.utils.to_categorical(trainY, num_classes=10)
testY = tf.keras.utils.to_categorical(testY, num_classes=10)
```

```python
# Now check the label size....
testY[0]
```

## Build the Graph

```python
# Clear out tensorflow memory..
tf.keras.backend.clear_session()

# Initialize Sequential model.  
model = tf.keras.models.Sequential()

# Reshape data from 2D to 1D -> 28x28 to 784.
model.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))                 # Layer = Mathematical model.  

# Normalize the data....
model.add(tf.keras.layers.BatchNormalization())
```

## Build the Graph
We will build a model with 4 hidden layers. Number of neurons in hidden layer will be 200, 100, 60 and 30 respectively. Both number of hidden layers and number of neurons in each hidden layer are hyperparameters *i.e* you can change these values to improve the model. Output of each neuron in hidden layer will be passed through an activation function.

## Build the Graph...

```python
# Compile the model.  

model.compile(optimizer='adam', loss='categorical_crossentropy', 
              metrics=['accuracy'])

# Or 
# sgd_optimizer = tf.keras.optimizers.SGD(lr=0.8, decay=0.001)
# model.compile(optimizer=sgd_optimiser, loss='mse’)   
```

## Review model

```python
model.summary()           
```

## 2) Weights = 784, Bias = 1. 
#### Neurons at Each Layers = 784-200-100-60-30-10.
#### So (784+1)*200 = 157000 Parameters at First Hidden Layers..
#### So (200+1)*100 = 20100 Parameters at Second Hidden Layers..
#### So (100+1)*60  =  6060 Parameters at Third Hidden Layers..
#### So (60+1)*30   =  1830 Parameters at Fourth Hidden Layers..
#### So (30+1)*10   =   310 Parameters at Output Layers..

## 1) batch_normalization Parameters.
#### Learnable Parameters = Trainable params = Scale & Shift Of Each Features = 1568 Parameters.
#### Non-Learnable Parameters = Non-trainable params = Mean & Variance Of Each Features = 1568 Parameters.

## Train the model

```python
mckpt = tf.keras.callbacks.ModelCheckpoint('./mnist_v1.h5', 
                                           monitor='val_acc', save_best_only=True)
```

```python
model.fit(trainX,trainY,          
          validation_data=(testX,testY),
          epochs=5,
          batch_size=32, callbacks=[mckpt])

# batch_size = All observations sent to Network per Epoch in defined batches.
# No. Of Gradient Descent Operations = (No. Of Observations/batch_size)*No. Of Epoch = (60000/32)*5 = 9375....
```
